**Abstract class example**

An abstract class is a class that cannot be instantiated. However, we can create classes that inherit from an abstract class.

An abstract method contains only method signature. The inheriting classes must implement the abstract method. An abstract class may or may not contains abstract methods.

Python does not support abstract class directly but provides a module called abc to define abstract classes.

**Example**
In the following example Employee class is defined as abstract so we cannot create an object of Employee class.

The child classes inheriting the Employee class must implement the abstract methods otherwise we will get error "Can't instantiate abstract class FullTimeEmployee with abstract methods get_number_of_work_hours"

```python
from abc import ABC, abstractmethod 
#Defining the abstract class
class Employee(ABC):
  
  @abstractmethod
  def get_number_of_work_hours(self):
    pass
  
  @abstractmethod
  def get_salary(self):
    pass

#Defining FullTimeEmployee child class which inherits Employee class
class FullTimeEmployee(Employee):
  
  def __init__(self,salary):
    self.salary = salary
  
  #Implementing the abstract method
  def get_number_of_work_hours(self):
    print("Full time employee work hours is 45")
  def get_salary(self):
    print("Full time employee salary is "+str(self.salary))

#Defining Contractor child class which inherits Employee class
class Contractor(Employee):
  
  def __init__(self,hourly_salary):
    self.hourly_salary = hourly_salary
  
  #Implementing the abstract method
  def get_number_of_work_hours(self):
    print("Contract employee work hours is 40")
  def get_salary(self):
    print("Full time employee salary is "+str(40*self.hourly_salary))

#creating object of the child class

FTE = FullTimeEmployee(10000)
FTE.get_number_of_work_hours()
FTE.get_salary()

contractor = Contractor(500)
contractor.get_number_of_work_hours()
contractor.get_salary()


```

**Multiple Inheritance**

When a child class inherits more than one base class it is called as multiple inheritance. The following example illustrates the multiple inheritance.

**Example1**

```python
#Defining the Shapes base class

class Shapes():
  def display(self):
    print("Shape class")
  def shape_charactertistics(self):
    print("All shapes has dimensions")

#Defining the Polygon base class

class Polygon():
  def display(self):
    print("Polygon class")
  def number_of_sides(self):
    print("Polygon has more than 1 side")

#Defining triangle class which inherits Shapes and Polygon class

class Triangle(Shapes,Polygon):
  #overriding base class method
  def display(self):
    print("Triangle class")
  def triangle_sides(self):
    print("A triangle has 3 sides")

t = Triangle()
t.display() 
t.triangle_sides()
```

**Decorator in python**

Decorators are used to modify the behaviour of function or class. In Decorators, functions are taken as the argument into another function and then called inside the wrapper function.

```python
def welcome_decorator(function_name):
  def welcome():
    print("Welcome to inner function")
    function_name()
  return welcome

def calling_function():
  print("Calling function")


obj = welcome_decorator(calling_function) #Function name passed as arguements
obj()


```
